```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\rmsin\Downloads\Customer Call List.xlsx")
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df["Last_Name"].str.strip()
#strip values from both sides 
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df
```

```python
df["Last_Name"]=df["Last_Name"].str.lstrip("/")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df
```

```python
df["Last_Name"].str.strip("123._/")
#this is another easy way to clean last name column in a single step
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('-','')
df["Phone_Number"]=df["Phone_Number"].str.replace('/','')
df["Phone_Number"]=df["Phone_Number"].str.replace('|','')
df["Phone_Number"]=df["Phone_Number"].str.replace('/','')
df["Phone_Number"]=df["Phone_Number"].str.replace('_','')
df
```

```python
str(df["Phone_Number"])
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
#using lambda
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand = True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df["Do_Not_Contact\t"] = df["Do_Not_Contact\t"].str.replace('Yes','Y')
df["Do_Not_Contact\t"] = df["Do_Not_Contact\t"].str.replace('No','N')
df
```

```python
df = df.fillna('')

df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
#another way to drop null values
#df = df.dropna(subset = "Phone_Number"), inplace = True))
```

```python
df = df.reset_index(drop = True)
df
```

```python
df = df.drop(columns = "Address")
df
```
